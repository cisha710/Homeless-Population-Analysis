```python
# Import dependencies
import pandas as pd
import os
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
# Select which columns to keep
columns_to_keep = ['State',
                   'Year',
                   'Overall Homeless - Non-Hispanic/Non-Latino',
                   'Overall Homeless - Hispanic/Latino',
                   'Overall Homeless - White',
                   'Overall Homeless - Black or African American',
                   'Overall Homeless - Asian',
                   'Overall Homeless - American Indian or Alaska Native',
                   'Overall Homeless - Native Hawaiian or Other Pacific Islander',
                   'Overall Homeless - Multiple Races',
                   'Sheltered Total Homeless - Non-Hispanic/Non-Latino',
                   'Sheltered Total Homeless - Hispanic/Latino',
                   'Sheltered Total Homeless - White',
                   'Sheltered Total Homeless - Black or African American',
                   'Sheltered Total Homeless - Asian',
                   'Sheltered Total Homeless - American Indian or Alaska Native',
                   'Sheltered Total Homeless - Native Hawaiian or Other Pacific Islander',
                   'Sheltered Total Homeless - Multiple Races',
                   'Unsheltered Homeless Individuals - Non-Hispanic/Non-Latino',
                   'Unsheltered Homeless Individuals - Hispanic/Latino',
                   'Unsheltered Homeless Individuals - White',
                   'Unsheltered Homeless Individuals - Black or African American',
                   'Unsheltered Homeless Individuals - Asian',
                   'Unsheltered Homeless Individuals - American Indian or Alaska Native',
                   'Unsheltered Homeless Individuals - Native Hawaiian or Other Pacific Islander',
                   'Unsheltered Homeless Individuals - Multiple Races']

```

```python
# Filter by rows (only CA and TX)
df_2020 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2020")
ca_tx_2020 = df_2020[(df_2020['State'] == 'CA') | (df_2020['State'] == 'TX')]
df_filtered_2020 = ca_tx_2020[columns_to_keep]

df_2019 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2019")
ca_tx_2019 = df_2019[(df_2019['State'] == 'CA') | (df_2019['State'] == 'TX')]
df_filtered_2019 = ca_tx_2019[columns_to_keep]

df_2018 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2018")
ca_tx_2018 = df_2018[(df_2018['State'] == 'CA') | (df_2018['State'] == 'TX')]
df_filtered_2018 = ca_tx_2018[columns_to_keep]

df_2017 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2017")
ca_tx_2017 = df_2017[(df_2017['State'] == 'CA') | (df_2017['State'] == 'TX')]
df_filtered_2017 = ca_tx_2017[columns_to_keep]

df_2016 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2016")
ca_tx_2016 = df_2016[(df_2016['State'] == 'CA') | (df_2016['State'] == 'TX')]
df_filtered_2016 = ca_tx_2016[columns_to_keep]

df_2015 = pd.read_excel('resources/demostate.xlsx', sheet_name = "2015")
ca_tx_2015 = df_2015[(df_2015['State'] == 'CA') | (df_2015['State'] == 'TX')]
df_filtered_2015 = ca_tx_2015[columns_to_keep]
```

```python
# Combined Texas and California data
combined_data = pd.concat([df_filtered_2020, 
                           df_filtered_2019,
                           df_filtered_2018, 
                           df_filtered_2017, 
                           df_filtered_2016,
                           df_filtered_2015], ignore_index=True)
# Sort combined_data by like state
custom_order = ['CA', 'TX']
combined_data['State'] = pd.Categorical(combined_data['State'], categories=custom_order, ordered=True)
combined_data = combined_data.sort_values('State')
combined_data['Year'] = combined_data['Year'].astype(int)
combined_data
```

```python
# Visual Analysis - Bar Graph 
```

```python
# Bar graph of the overall homeless population 
plt.figure(figsize=(15, 9))

plt.bar(combined_data['Year'], combined_data['Overall Homeless - Non-Hispanic/Non-Latino'], label='Non-Hispanic/Non-Latino', alpha=0.7)
plt.bar(combined_data['Year'], combined_data['Overall Homeless - Hispanic/Latino'], label='Hispanic/Latino', alpha=0.7)
plt.bar(combined_data['Year'], combined_data['Overall Homeless - White'], label='White', alpha=0.7)
plt.bar(combined_data['Year'], combined_data['Overall Homeless - Black or African American'], label='Black or African American', alpha=0.7)
plt.bar(combined_data['Year'], combined_data['Overall Homeless - Asian'], label='Asian', alpha=0.7)
plt.bar(combined_data['Year'], combined_data['Overall Homeless - American Indian or Alaska Native'], label='American Indian or Alaska Native', alpha=0.7)
plt.bar(combined_data['Year'], combined_data['Overall Homeless - Native Hawaiian or Other Pacific Islander'], label='Native Hawaiian or Other Pacific Islander', alpha=0.7)
plt.bar(combined_data['Year'], combined_data['Overall Homeless - Multiple Races'], label='Multiple Races', alpha=0.7)

plt.xlabel('Year', fontsize = 20)
plt.ylabel('Homeless Individuals', fontsize = 20)
plt.title('Overall Homeless Individuals in California and Texas', fontsize = 32)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(fontsize = 15, ncol=2)

plt.tight_layout()
plt.savefig('OverallHomelessCaTx.png')
plt.show()
```

To begin my visual data analysis I created a bargraph with the cleaned California and Texas dataset. The x-axis represents the years 2015 - 2010 in which racial data was collected, and the y-axis represents the total amount of overall homeless individuals in the states of California and Texas.

Based on the bar graph Non-Hispanic/Non-Latino (blue), White (green), and Black or African American (red) appear to be the three most affected races by overall homelessness. Whereas other groups such as Asian, Native Hawaiian or other Pacidic Islander, and multiple races appear to be less affected. One reason this that could explain this trend is that there are likely a lot more indiviuals in these states that identify as Non-Hispaic, White, or Black than individals that identify as Asian, Native Hawaiian, or as multiple races. Another thing to point out in this graph is there appears to be an increasing trend of overall homelessness in the three main race groups overtime.

```python
# Anova comparing the overall, sheltered, unsheltered non-hispanic/non-latino homeless indiviuals in TX
import pandas as pd
from scipy import stats
import statsmodels.api as sm

f_statistic, p_value = stats.f_oneway(combined_tx['Overall Homeless - Non-Hispanic/Non-Latino'], 
                                      combined_tx['Sheltered Total Homeless - Non-Hispanic/Non-Latino'], 
                                      combined_tx['Unsheltered Homeless Individuals - Non-Hispanic/Non-Latino'])

print(f'F-statistic: {f_statistic}')
print(f'p-value: {p_value}')
```

Ho = There are no significant differences among the homeless groups non-hispanic/non-latino individuals in the state of Texas. 
Ha = There is a significant difference among at least one of the homeless groups non-hispanic/non-latino individuals in the state of Texas.

Since the p-value < 0.05 there is evidence to reject the null hypothesis in favor of the alternative hypothesis. At least one of the homeless groups of non-hispanic/non-latino individuals in the state of Texas is significantly different than the others. A Tukey's test or Bonferroni correction can also be performed to identify which specific groups differ from each other.
